```python
import search_tool
from IPython.display import HTML
```

### Graph Initialization
Define a graph by declaring a list of nodes and edges, as well as a start node and a goal node for the search functions. Each edge is a tuple consisting of two vertices and the optional edge cost. Nodes can also be defined as a tuple consisting of a name for the node and a weight/heuristic for the node.

Graphs can be viewed as a networkx graph or displayed as an adjacency list.

```python
nodes = ['A', 'B', 'C', 'D']
edges = [('A', 'B'), ('B', 'C'), ('B', 'D')]
start, goal = 'A', 'C'
graph = search_tool.Graph(nodes, edges, start, goal)
graph.visualize_graph()
graph.print_adj_list()
```

```python
nodes = ['A', 'B', 'C', 'D']
edges = [('A', 'B', 3), ('B', 'C', 5), ('B', 'D', 2)]
start, goal = 'A', 'C'
graph_cost = search_tool.Graph(nodes, edges, start, goal, True)
graph_cost.visualize_graph()
graph_cost.print_adj_list()
```

### Search Functionality
The Search Tool package includes functions to perform the following search algorithms on a graph:
* Breadth First
* Depth First
* A-Star
* Iterative Deepening
* Beam Search
* Hill-Climbing

### Breadth First Search
This function performs a Breadth-First Search on the provided graph and returns the frontier at each iteration of the search, the explored nodes at each iteration of the search, and the final path from the start node to the goal node.

```python
bfs_frontier, bfs_explored, bfs_path = search_tool.BFS(graph)
```

### Search Visualization
The search tool package provides animations of the frontier, explored nodes, and solution path found by a search algorithm. As the search is executed, the node colors are updated to show what nodes have been explored and expanded at each iteration.
* Cyan - nodes that have not yet been explored during search
* Yellow - nodes that currently make up the frontier
* Purple - nodes that have been explored and expanded
* Green - nodes that are part of the path found during search

```python
bfs_animation = search_tool.animate_search(graph, bfs_frontier, bfs_explored, bfs_path)
search_tool.save_animation(bfs_animation, "BFS-example.mp4")
HTML(bfs_animation.to_html5_video())
```

### Depth First Search
This function performs a Depth-First Search on the provided graph and returns the frontier at each iteration of the search, the explored nodes at each iteration of the search, and the final path from the start node to the goal node.

```python
dfs_frontier, dfs_explored, dfs_path = search_tool.DFS(graph)
dfs_animation = search_tool.animate_search(graph, dfs_frontier, dfs_explored, dfs_path)
search_tool.save_animation(dfs_animation, "DFS-example.mp4")
HTML(dfs_animation.to_html5_video())
```

### A-Star Search
This function executes the A-Star search algorithm on a graph in which heuristic costs are provided for each vertex during intialization. If a cost is not provided for a node, it is assumed to be zero. 

```python
a_star_frontier, a_star_explored, a_star_path = search_tool.a_star_search(graph_cost)
a_star_animation = search_tool.animate_search(graph, a_star_frontier, a_star_explored, a_star_path)
search_tool.save_animation(a_star_animation, "A-Star-example.mp4")
HTML(a_star_animation.to_html5_video())
```

### Iterative Deepening

```python
id_solution, id_frontier, id_explored, id_path = search_tool.interative_deepening(graph_cost)
id_animation = search_tool.animate_search(graph_cost, id_frontier, id_explored, id_path)
search_tool.save_animation(id_animation, "Iterative-Deepening-example.mp4")
HTML(id_animation.to_html5_video())
```

### Beam Search

```python
beam_search_path = search_tool.beam_search(graph_cost)
print("Goal Path (beam search):")
print(' -> '.join(beam_search_path))
```

```python
hc_solution, hc_frontier, hc_explored, hc_path = search_tool.hill_climbing(graph_cost)
hc_animation = search_tool.animate_search(graph_cost, hc_frontier, hc_explored, hc_path)
search_tool.save_animation(hc_animation, "Hill-Climbing-example.mp4")
HTML(hc_animation.to_html5_video())
```

### Search Comparison
The Benchmarking class includes functions to perform a comparison of each search algorithm.The benchmark function provides a comparison of the time complexity.

```python
search_inputs = [[graph_cost], [graph_cost], [graph_cost], [graph_cost], [graph_cost], [graph_cost]]
comparison = search_tool.Benchmarking(search_tool.BFS, search_tool.DFS, search_tool.a_star_search, 
                                      search_tool.interative_deepening, search_tool.beam_search, 
                                      search_tool.hill_climbing, search_inputs)
comparison.benchmark()
```
